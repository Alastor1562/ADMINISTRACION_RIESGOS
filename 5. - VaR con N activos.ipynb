```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
tickers = ['SOXX', 'IBB']
prices = yf.download(tickers, start='2025-02-07', end='2026-02-10', progress=False)['Close']
prices
```

**Caso 1:** SOXX Long, IBB Long

```python
returns = prices.pct_change().dropna()

def portfolio_var(returns, weights, confidence=0.95):
    weights = pd.Series(weights, dtype=float)
    aligned = returns[weights.index]
    port_ret = aligned @ weights.values
    alpha = (1 - confidence) * 100
    return -np.percentile(port_ret, alpha)

weights = {"IBB": 6/16, "SOXX": 10/16}
VaR_95 = portfolio_var(returns, weights, confidence=0.95)
VaR_95 * 100
```

```python
VaR_95 * 16000000
```

**Caso 2:** SOXX Long, IBB Short

```python
weights_short = {"IBB": -6/16, "SOXX": 10/16}
```

```python
VaR_95_short = portfolio_var(returns, weights_short, confidence=0.95)
VaR_95_short * 100
```

```python
VaR_95_short * 16000000
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
```

```python
ask = 5.79
bid = 3.78

mid = (ask + bid) / 2
spread_pct = (ask - bid) / mid
print(f"El spread porcentual es: {spread_pct:.2%}")
```

```python
df = pd.read_excel(
        'Data_C3.xlsx',
        header=2
    )

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
df['spread_pct'] = (df['Ask'] - df['Bid']) / ((df['Ask'] + df['Bid']) / 2)
mean_spread = df['spread_pct'].mean()
std_spread = df['spread_pct'].std()

returns = df['Last'].pct_change().dropna()

mean_spread, std_spread
```

```python
var = -np.percentile(returns, 1)
print(f"El VaR al 99% de confianza es: {var:.2%}")
```

```python
# Calculo del L-VaR paramétrico
z = norm.ppf((1 + 0.99) / 2) # valor z para 99% de confianza

lvar = var + 0.5 * (mean_spread + z * std_spread)
print(f"El L-VaR al 99% de confianza es: {lvar:.2%}")
```

```python
# Calculo del L-VaR histórico
lvar_hist = var - np.percentile(df['spread_pct'], 1)
lvar_hist
```
